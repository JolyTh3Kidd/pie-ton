```python
import matplotlib.pyplot as plot

names = ['Kurmet', 'Duman', 'Alan', 'Ilyas', 'Almansur']
scores = [85, 95, 70, 75, 88]

plot.figure(figsize=(8, 5))
plot.plot(names, scores, marker='o', linestyle='-', color='b', label='Score Trend')

plot.xlabel('Student Names')
plot.ylabel('Scores')
plot.title('Student Scores (Line Plot)')
plot.legend()
plot.grid(True)
plot.show()
```

```python
plot.figure(figsize=(8, 5))
plot.bar(names, scores, color='green')

plot.xlabel('Student Names')
plot.ylabel('Scores')
plot.title('Student Scores')
plot.show()
```
