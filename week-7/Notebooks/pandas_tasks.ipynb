```python
import pandas as pd

data = {
    'Name': ['Kurmet', 'Duman', 'Alan', 'Ilyas', 'Almansur'],
    'Age': [20, 21, 20, 18, 18],
    'Score': [85, 95, 70, 75, 88]
}

dataframe = pd.DataFrame(data)
print("Initial DataFrame:")
print(dataframe)
```

```python
print("\nFirst 3 Rows:")
print(dataframe.head(3))

print("\nSummary Statistics:")
print(dataframe.describe())

average = dataframe['Score'].mean()
print(f"\nAverage Score: {average}")
```

```python
older_than_20 = dataframe[dataframe['Age'] > 20]
print("\nStudents older than 20:")
print(older_than_20)

high_scores = dataframe[dataframe['Score'] > 80]
print("\nStudents with score > 80:")
print(high_scores)
```
