```python
import numpy as np

array_1d = np.arange(1, 11)
print("1D Array:", array_1d)
array_2d = np.random.randint(0, 100, (3, 3))
print("\n2D Array (Random):\n", array_2d)
```

```python
print("\nSum:", array_1d.sum())
print("Mean:", array_1d.mean())
print("Max:", array_1d.max())
array_multiplied = array_1d * 2
print("\nArray multiplied by 2:", array_multiplied)
filtered_array = array_1d.copy()
filtered_array[filtered_array > 5] = 0
print("\nReplaced values > 5 with 0:", filtered_array)
```
